**Import Library**

```python
import matplotlib.pyplot as plt
```

**Dataset**

```python
# Buat data sample
months = ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Des']
product_A_sales = [10,20,15,25,30,45,40,50,60,55,65,70]
product_B_sales = [5,10,8,15,18,20,22,30,25,35,40,45]
```

**Single Line Plot**

```python
# line plot 1 garis, sytle grafik default
plt.plot(months, product_A_sales, label='Product A')
plt.title('Penjualan Produk A per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

**Multiple Line Plot (Line 1 - product A)**

```python
# kustomisasi line plot
plt.plot(months, product_A_sales, label='Product A', linestyle="--", marker='o')
#
plt.plot(months, product_B_sales, label='Product B', linestyle="-", marker='x')

plt.title('Penjualan Produk per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

**Jenis Garis untuk Menunjukan Tren**

```python
# Data sample tambahan
product_C_sales = [18,2,25,28,32,38,42,45,48,52,56,60]
product_D_sales = [7,8,9,14,21,25,30,35,40,51,48,55]

# Plot product A
plt.plot(months, product_A_sales, label='Product A', linestyle="--", color='blue')

# B
plt.plot(months, product_B_sales, label='Product B', linestyle="-", color='red')

# C
plt.plot(months, product_C_sales, label='Product C', linestyle=":", color='green')

# D
plt.plot(months, product_D_sales, label='Product D', linestyle="-.", color='purple')
plt.title('Penjualan Produk per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

**Subplot pada Line Chart**

```python
# sublot, 2 baris 1 kolom, panjang 10 inc, tinggi 8 inc
fig, axs = plt.subplots(2, 1, figsize=(10,8))

# plot pertama untuk product C
axs[0].plot(months, product_C_sales, label='Product C', color='green', marker='d')
axs[0].set_title('Penjualan Produk C per Bulan')
axs[0].set_xlabel('Bulan')
axs[0].set_ylabel('Penjualan')
axs[0].legend()
axs[0].grid(True)

# plot pertama untuk product D
axs[1].plot(months, product_D_sales, label='Product D', color='blue', marker='s')
axs[1].set_title('Penjualan Produk D per Bulan')
axs[1].set_xlabel('Bulan')
axs[1].set_ylabel('Penjualan')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
